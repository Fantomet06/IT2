# Oppgave 1

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def f(x): return x**3
def df(x): return 3*x**2
def ddf(x): return 6*x
```

```python
x = np.linspace(-2, 2, 100)

plt.plot(x, f(x), 'b',label="f(x) = $x^3$") # f(x)
plt.plot(x, df(x), 'r', label="f'(x) = $3x^2$") # f'(x)
plt.plot(x, ddf(x), 'g', label="f''(x) = 6x") # f''(x)

plt.xlabel('x')
plt.ylabel("y")
plt.title("Grafene til f(x), f'(x) og f''(x)")

plt.axhline(0, color='black')
plt.axvline(0, color='black')

plt.legend()
plt.show()
```

```python
fig, axs = plt.subplots(3, 1, figsize=(8, 12))

axs[0].plot(x, f(x), 'b', label="f(x) = $x^3$") # f(x)
axs[0].set_title("Graf av f(x) = $x^3$")

axs[1].plot(x, df(x), 'r', label="f'(x) = $3x^2$") # f'(x)
axs[1].set_title("Graf av f'(x) = $3x^2$")

axs[2].plot(x, ddf(x), 'g', label="f''(x) = 6x") # f''(x)
axs[2].set_title("Graf av f''(x) = 6x")

for i in range(3):
    axs[i].axhline(0, color='black', linestyle='dashed')
    axs[i].axvline(0, color='black', linestyle='dashed')
    axs[i].legend()

plt.tight_layout()
plt.show()
```

# Oppgave 2

```python
import random
import matplotlib.pyplot as plt

def kaste_terning_simulator_3000(antall_kast: int) -> list:
    resultater = [0]*6 # index representerer tallet
    
    for i in range(antall_kast):
        kast = random.randint(1, 6)
        resultater[kast-1] += 1
    
    return resultater

# c. Simulate 100000 dice rolls and plot the graph
antall_kast = kaste_terning_simulator_3000(100000)
plt.scatter([1,2,3,4,5,6], antall_kast)
plt.ylim(min(antall_kast)-100, max(antall_kast) + 100)
plt.xlabel('Side')
plt.ylabel('Antall ganger')
plt.title('Terningskastsimulator 3000')
plt.show()
```
