```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(-2, 2, 100)
y_6x = 6*x
y_3x2 = 3*x**2
y_x3 = x**3

plt.plot(x, y_x3, 'b',label="f(x) = $x^3$")
plt.plot(x, y_3x2, 'r', label="f'(x) = $3x^2$")
plt.plot(x, y_6x, 'g', label="f''(x) = 6x")


plt.xlabel('x')
plt.ylabel("y")
plt.title("Grafene til f(x), f'(x) og f''(x)")

plt.axhline(0, color='black')
plt.axvline(0, color='black')

plt.legend()
plt.show()
```

```python
fig, axs = plt.subplots(3, 1, figsize=(8, 12))

axs[0].plot(x, y_x3, 'b', label="f(x) = $x^3$")
axs[0].set_title("Graf av f(x) = $x^3$")
axs[0].legend()

axs[1].plot(x, y_3x2, 'r', label="f'(x) = $3x^2$")
axs[1].set_title("Graf av f'(x) = $3x^2$")
axs[1].legend()

axs[2].plot(x, y_6x, 'g', label="f''(x) = 6x")
axs[2].set_title("Graf av f''(x) = 6x")
axs[2].legend()

for i in range(3):
    axs[i].axhline(0, color='black', linestyle='dashed')
    axs[i].axvline(0, color='black', linestyle='dashed')

plt.tight_layout()
plt.show()
```

# Oppgave 2

```python
import random
import matplotlib.pyplot as plt

def kaste_terning_simulator_3000(antall_kast: int) -> list:
    resultater = [0]*6
    
    for i in range(antall_kast):
        kast = random.randint(1, 6)
        resultater[kast-1] += 1
    
    return resultater

# c. Simulate 100000 dice rolls and plot the graph
antall_kast = kaste_terning_simulator_3000(100000)
plt.bar([1,2,3,4,5,6], antall_kast)
plt.ylim(min(antall_kast)-100, max(antall_kast) + 100)
plt.xlabel('Side')
plt.ylabel('Antall ganger')
plt.title('Terningskastsimulator 3000')
plt.show()
```
