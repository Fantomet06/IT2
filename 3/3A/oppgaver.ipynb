```python
import matplotlib.pyplot as plt
import pandas as pd
```

#### Oppgave 3.11

```python
# oppgave 3.11

data = [
    [1985, 71, 50],
    [1989, 63, 37],
    [1993, 67, 28],
    [1997, 65, 23],
    [2001, 43, 38],
    [2005, 61, 38],
    [2009, 64, 41],
    [2013, 55, 48],
    [2017, 49, 45],
    [2021, 48, 36]
]

aar = [row[0] for row in data]
values1 = [row[1] for row in data]
values2 = [row[2] for row in data]

bar1 = plt.bar(aar, values1, width=2, color='b')
bar2 = plt.bar(aar, values2, width=2, color='r')
plt.xticks(aar)
plt.xlabel('År')
plt.ylabel('Representanter')
plt.legend((bar1, bar2), ('Høyre', 'Ap'))
plt.show()

```

```python
# Oppgave 3.11 men med pandas

plot = pd.DataFrame(data, columns=['År', 'Høyre', 'Ap'])
plot.plot.bar(x='År', y=['Høyre', 'Ap'], color=['b', 'r'])
```
