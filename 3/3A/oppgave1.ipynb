```python
import matplotlib.pyplot as plt
```

```python
datasett1 = "./datasett1.txt"
datasett2= "./datasett2.txt"
```

# Datasett 1: Fordeling av karakterer

```python
antall = [0]*6
karakterer = [x for x in range(1,7)]

with open(datasett1) as f:
    lines = f.readlines()
    lines = lines[0].split(",")
    for i in lines:
        antall[int(i)-1] += 1

plt.bar(karakterer, antall)
plt.show()
```

# Datasett 2: Tid brukt p√• 60m

```python
antall = [0]*6


with open(datasett2) as f:
    data = f.readlines()
    data = data[0].split(", ")
    data = [float(x) for x in data]

a, b = round(min(data),0), round(max(data), 0)+2

plt.hist (data, bins = [x for x in range(int(a),int(b))],edgecolor = "black")
plt.title ("Elevene p√• 60 m")
plt.ylabel ("Antall")
plt.xlabel ("Sekunder p√• 60m")
plt.grid (axis = "y")
plt.show()
```
